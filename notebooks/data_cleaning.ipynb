```python
from IPython import get_ipython
get_ipython().run_line_magic('cd', '..')

from data.features_abstractions import features_abstractions
```

```python
for feature, abstractions in features_abstractions.items():
    for abstraction, subcategories in abstractions.items():
        df[feature] = df[feature].replace(to_replace=subcategories, value=abstraction)
```

```python
df.head()
```

```python
for feature in features_abstractions.keys():
    print(f'Quantidade de categorias na coluna {feature}: {df[feature].nunique()}')
```

```python
for feature, abstractions in features_abstractions.items():
    print(f'\n{feature}:\n')
    all_abstractions = []
    all_subcategories = []

    for abstraction, subcategories in abstractions.items():
        all_abstractions.append(abstraction)
        all_subcategories.append(subcategories)

    for category in df[feature].unique():
        if category not in all_abstractions and category not in all_subcategories:
            print(f'"{category}"')
    
print('\nFinished!')

# identificar se há alguma categoria do dataset que não está mapeada no dicionário
```

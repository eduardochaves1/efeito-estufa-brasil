```python
emissao_bunker = df.query('tipo_emissao == "Bunker"')
emissao_bunker.head()
```

```python
for feature in emissao_bunker.columns:
    print(f'Categorias da Coluna {feature}:\n')
    print(emissao_bunker[feature].value_counts())
    print('\n\n')
```

```python
df.sort_values('emissao')
```

```python
df_not_emission = df.query('emissao < 0')
df_not_emission.head()
```

```python
df_not_emission['tipo_emissao'].value_counts()
```

```python
for col in df_not_emission.columns:
  if 'nivel_' in col:
    print(f'\n{col}:')
    print(df_not_emission[col].unique())
```

```python
df_emissions = df.drop(df_not_emission.index)
df_emissions.head()
```

```python
for col in df_emissions.columns:
  if 'nivel_' in col:
    print(f'\n{col}:')
    print(df_emissions[col].unique())
```
